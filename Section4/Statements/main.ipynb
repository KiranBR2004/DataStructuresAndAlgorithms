## Condition Statements In Python (if, else, elif)



```python
## if statement
age=18
if(age<20):
    print("You are eligiable to vote in election")
```

```python
age>=18
```

```python
## else
## The else Statement Executes in Block of Code if the condition in the if statement is false

age=16
if(age>=18):
    print("you are eligible to voting in the election")
else:
    print("you are a minor")
```

```python
## elif
## the elif statement allows you to check multiple condition it stands for "else if"

age=20
if(age<10):
    print("you are a child")
elif(age<18):
    print("you are a teenager")
else:
    print("you are a adult")
```

```python
## nested condition statement
## you can place one or more if, elif, else statement inside the another if, elif, else statement
## number even, odd, positive,nagitive

num=int(input("enter a number"))

if(num>=0):
    print("the number is positive")
    if(num%2==0):
        print("the number is even")
    else:
        print("the number is odd")
else:
    print("the numer is zero or nagitive")
```
