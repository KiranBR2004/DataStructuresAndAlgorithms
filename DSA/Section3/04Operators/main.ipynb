## OPerators In Python

```python
## Arthmatic Operation

a=20
b=10

add_result=a+b #Addition
sub_result=a- #Substraction
mul_result=a*b #Multiplication
div_result=a/b #Division
floor_div_result=a//b #Floor_Division
modulus_result=a%b #Modulus operation
exponent_result=a**b #Exponentiation

print(add_result)
print(sub_result)
print(mul_result)
print(div_result)
print(floor_div_result)
print(modulus_result)
print(exponent_result)
```

## Comparison Operation

```python
# Comparison Operation
# == Equal To
a=20
b=20

a==b
```

```python
# grater than >

a=10
b=20
a>b
```

```python
# lesser than <

print(a<b)
```

```python
# grater than or equal >=

num1=25
num2=25
num1>=num2
```

## Logical Operater

```python
# And Not OR
X=True
Y=True

result = X and Y
print(result)
```

```python
X=False
Y=True

result = X and Y
print(result)
```

```python
# OR
X=True
Y=True

result = X or Y
print(result)
```

```python
X=False
Y=False

result = X or Y
print(result)
```

```python
# NOT Operation

X=False
not X
```
